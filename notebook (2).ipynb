How will we do it? We will create a function that will help us calculate profit from a set of selected oil wells and the models' predictions:

````python
def calculate_profit(predictions, targets):
    # Reset indices to ensure alignment
    predictions = predictions.reset_index(drop=True)
    targets = targets.reset_index(drop=True)

    # Get indices of top 200 predicted wells
    top_200_idx = predictions.sort_values(ascending=False).head(200).index

    # Select actual target values for those wells
    selected_targets = targets.loc[top_200_idx]

    # Calculate profit
    total_product = selected_targets.sum()
    revenue = total_product * revenue_per_thousand
    profit = revenue - budget

    return profit
````

According to our findings, the most profitable would be Region 0. However, The prediction to use the top 200 wells from this region sounds unreliable due to the high RMSE. Which means it is being overestimating its profit and on actual practice could be catastrophic.

Not to worry! That is why we will now calculate risks and profit for each region to select the best way to go for the company! We will start with creating a function to bootstrap with 1000 samples to find the distribution of profit.

````python
def bootstrap_profit(predictions, targets, n_simulations=1000):
    profits = []
    rng = np.random.RandomState(12345)  # Controls reproducibility
    # Generate a unique seed for each iteration using the main RNG
    state = rng.randint(0, 10**6)
    for _ in range(n_simulations):
        sample = predictions.sample(n=500, replace=True, random_state=state)
        sample_targets = targets.loc[sample.index]
        profit = calculate_profit(sample, sample_targets)
        profits.append(profit)

    return profits
````

Now, to analyze the results:

````python
def analyze_bootstrap_results(profits, region_name):
    profits = pd.Series(profits)
    mean_profit = profits.mean()
    ci_lower = profits.quantile(0.025)
    ci_upper = profits.quantile(0.975)
    risk = (profits < 0).mean() * 100 

    print(f"\n--- {region_name} ---")
    print(f"Mean profit: ${mean_profit:,.2f}")
    print(f"95% CI: ${ci_lower:,.2f} to ${ci_upper:,.2f}")
    print(f"Risk of loss: {risk:.2f}%")

analyze_bootstrap_results(profits_0, "Region 0")
analyze_bootstrap_results(profits_1, "Region 1")
analyze_bootstrap_results(profits_2, "Region 2")
````
